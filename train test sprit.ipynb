# train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['overall_score']),df['overall_score'],
                                                test_size=0.2)
```

# Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
# Ordinalencoding -> parent_education,travel_time
oe = OrdinalEncoder(categories=[
    ['no formal','high school','diploma','graduate','post graduate','phd'],
    ['<15 min','15-30 min','30-60 min','>60 min']])
X_train[['parent_education','travel_time']] = oe.fit_transform(X_train[['parent_education','travel_time']])
# also the test data
X_test[['parent_education','travel_time']] = oe.fit_transform(X_test[['parent_education','travel_time']])
```
